### Flippausvinkkejä/kehitettävää:
•	Vaikka ostoskorianalyysiin mentiinkin demoissa syvemmälle, olisi ollut hyvä saada pieni demo myös ostoskorianalyysistä luennolla. 

### Koodia

```python
#Tutkitaan varustelekan yritysesittelysivun tekstiä
import bs4
import requests
import nltk

page_html=requests.get("https://www.varusteleka.fi/fi/page/yritysesittely/19791")
page_parsed=bs4.BeautifulSoup(page_html.content,'html.parser')

text = ""
counts = {}
stopwords = [".", ":", "-"]

for word in page_parsed.get_text().split():
    word = word.replace(",", "")
    if any(x in word for x in stopwords) == False and len(word) > 3 and any(char.isdigit() for char in word) == False:
        word = "" + word.lower()
        if word in counts.keys():
            counts[word] = counts[word] + 1
        else:
            counts[word] = 1
        text += " " + word

print(sorted(counts, key=counts.get, reverse=True)[:3]) #3 eniten esiintyvää sanaa dokumentissa

tokens = nltk.word_tokenize(text) #NLTK word_tokenize ei tue suomenkieltä, eli tässä pitäisi käyttää jotain suomenkieltä tuntevaa kirjastoa
res = nltk.Text(tokens)
res.dispersion_plot(["lekalainen", "vuodesta", "varustelekan"])
#Vaikkei tokenisointa voitu tehdä niin saatiin silti selville ilmiö jossa sanat 'vuodesta' ja 'lekalainen' näyttävät toistuvan samassa paikassa dokumenttia
```

### Koodia:

```python
#Tutkitaan lineaarisen regression avulla ja visualisoimalla hinnan suhdetta makuuhuoneisiin Airbnb-datan avulla
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

LISTINGSFILE = "http://data.insideairbnb.com/italy/lombardy/milan/2021-12-15/data/listings.csv.gz"
cols = ['price',
        'bedrooms']
df = pd.read_csv(LISTINGSFILE, usecols=cols, compression='infer')

#Jalostetaan dataa
df['price'] = df['price'].replace('[\$,)]','',  \
        regex=True).replace('[(]','-', regex=True).astype(float)
df = df[df['bedrooms'].notna()]
df = df[df['price'] < 400]

x = df['bedrooms'].values
y = df['price'].values
plt.plot(x,y, 'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
#Voidaan huomata hinnan nousevan mitä enemmän makuuhuoneita asunnoissa on, joka käy hyvin järkeen
```
