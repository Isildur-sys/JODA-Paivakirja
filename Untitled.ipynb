```python
#Normalisoidaan ja siivotaan laadullista dataa
import pandas as pd
import numpy as np

LISTINGSFILE = "http://data.insideairbnb.com/italy/lombardy/milan/2021-12-15/data/listings.csv.gz"

cols = ['host_response_time']
df = pd.read_csv(LISTINGSFILE, usecols=cols, compression='infer')

df['host_response_time'] = df['host_response_time'].fillna('N/A')
dummies = pd.get_dummies(df.host_response_time)
df = pd.concat((df.drop(['host_response_time'], axis = 1), dummies.astype(int)), axis = 1)
df.describe()

xMean = np.mean(df, axis=0)
xDev = np.std(df, axis=0)
xNorm = (df - xMean) / xDev
xNorm
```
